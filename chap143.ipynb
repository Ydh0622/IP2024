#### Translate

```python
img = cv2.imread('DOGe.jpg',0)
rows,cols = img.shape[:2]
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
```

#### Rotate

```python
img = cv2.imread('DOGe.jpg')
rows,cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/4,rows/4),43,1)
dst = cv2.warpAffine(img,M,(cols,rows))
```

```python
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##### Affine Transformation

```python
img = cv2.imread('DOGe.png')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##### Perspective Transformation

```python
img = cv2.imread('Notebook.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[[170, 190],[760, 196],[808, 1096],[100, 1080]]])
pts2 = np.float32([[0,0],[400,0],[400,600],[0,600]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(400,600))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```
